# 18. Other libraries

There are many alternatives to the solutions shown during the course. Depending on your field and background you might want to search for another solution. Generally however libraries work mostly like seaborn i.e. you pass a data element to a plotting function and then specify which "aspect" of the data you want to plot and use for color, marker etc. aesthetics.

```python
import pandas as pd
import numpy as np

diams = pd.read_csv('https://raw.githubusercontent.com/vincentarelbundock/Rdatasets/master/csv/Ecdat/Diamond.csv')
```

## Plotnine

[Plotnine](https://plotnine.readthedocs.io/en/stable/index.html) is one alternative to seaborn. It is in fact an almost perfect copy of ggplot, so if you are familiar with that R package but want to "unify" your pipeline in Python, it could be a good alternative.

The functions have mostly the same name as in ggplot, and the *assembling* of plots works in the same way. Two examples here:

```python
from plotnine import ggplot, geom_point, aes, stat_smooth, facet_wrap, geom_boxplot
```

```python
fig = (ggplot(diams, aes('carat', 'price', color='colour'))
 + geom_point()
 + stat_smooth(method='lm')
);
fig
```

```python
fig = (
    ggplot(diams)
    + geom_boxplot(aes(x='colour', y='price'))
)
fig
```

## Altair

Altair is another library that works similarly to seaborn. It is based on an entirely different ecosystem (non Matplotlib) but offeres interesting possibilities in terms of interactivity.

Here an example as well:

```python
import altair as alt

alt.Chart(diams).mark_point().encode(
    x='price',
    y='carat',
    color='colour',
).interactive()
```

## Plotly

[Plotly](https://plotly.com/) is also a very popular library. It is in particular widely used to create interactive dashboards running on a server. The principles are again the same as for seaborn plotting:

```python
import plotly.express as px

fig = px.scatter(diams, x="carat", y="price", color="colour")
fig.show()
```
