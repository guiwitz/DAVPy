````
import pandas as pd
````

1. Load a first DataFrame that contains information about covid19 vaccination locations: https://github.com/owid/covid-19-data/blob/master/public/data/vaccinations/locations.csv 

````
locations = pd.read_csv('https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/vaccinations/locations.csv')
````

2. Load a second DataFrame which contains daily information about vaccination for every country: https://github.com/owid/covid-19-data/blob/master/public/data/vaccinations/vaccinations.csv

````
vaccinations = pd.read_csv('https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/vaccinations/vaccinations.csv')
````

3. Create a sub-dataframe of the table in (1) that only contains the fields ```location``` and ```vaccines```.

````
sub_df = locations[['location', 'vaccines']]
````

````
sub_df.head(3)
````

4. Merge the two tables using the ```location``` field as key for merging. Use once ```left``` merge and once ```right``` marge.

````
left_merge = pd.merge(vaccinations, sub_df, on='location', how='left')
right_merge = pd.merge(vaccinations, sub_df, on='location', how='right')
````

5. Do you see a difference in the result?


````
left_merge.shape
````

````
right_merge.shape
````

The two tables don't contains exactly the same locations and therefore depending if we merge based on left or right, we don't get the same result.

5b. Bonus question: knowing that you can use the ```.unique()``` method on columns, and that there's a function called ```np.isin``` that compares two lists, can you figure out which country is missing ?

````
left_locations = left_merge.location.unique()
````

````
right_locations = right_merge.location.unique()
````

````
import numpy as np
is_present = np.isin(left_merge.location.unique(),right_merge.location.unique())
````

```is_presnt``` is a boolean array that tells us if a given element of ```left_locations``` is present in ```right_locations```. We want to extract the ```False``` values, so we need to do indexing with a "negated" version of ```is_present```:

````
left_locations[~is_present]
````
